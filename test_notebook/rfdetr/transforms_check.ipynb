```python
from hq_det.trainer import DefaultAugmentation
from hq_det.tools.train_rfdetr import MyTrainer
from hq_det.trainer import HQTrainerArguments
from raw_utils import create_dataloaders
from utils import TransformTracker
```

```python
dataset_path = "/root/autodl-tmp/chengdu_v2_dataset"
```

```python
raw_data_loader_train, raw_data_loader_val = create_dataloaders(dataset_dir="/root/autodl-tmp/chengdu_v2_dataset")
```

```python
trainer = MyTrainer(
    HQTrainerArguments(
        data_path=dataset_path,   # 数据集路径
        num_epoches=10,  # 训练轮数
        warmup_epochs=5,  # 预热轮数
        num_data_workers=4,  # 数据加载线程数
        lr0=1e-4,  # 初始学习率
        lr_min=1e-5,  # 最小学习率
        batch_size=16,  # 批量大小
        device='cuda',  # 设备
        checkpoint_path='./output',  # 检查点路径
        output_path='./output',  # 输出路径
        checkpoint_interval=1,  # 检查点间隔
        image_size=1024,  # 图像大小
        gradient_update_interval=2,  # 梯度更新间隔
        model_argument={
            "model": "/root/autodl-tmp/model/rfdetr/rf-detr-base.pth",  # 加载的模型路径
            "model_type": "base",  # 模型类型
            "lr_encoder": 1.5e-4,  # 编码器学习率
            "lr_component_decay": 0.7,  # 组件衰减率
        },
        eval_class_names=[],   # 评估类别名称
    )
)
```

```python
train_transforms = trainer.build_transforms(aug=True)
transform_tracker = TransformTracker(trainer.args.data_path)
res = transform_tracker("train", 1275, train_transforms.transforms)
transform_tracker.visualize_transforms(res)
```
